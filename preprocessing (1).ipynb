```python
# SimpleIMPUTER, only works if you have only numerical values
# won't work for categorical values??
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer

mydata = {'one': pd.Series([2, 5, 9], index = ['a', 'b', 'c']),
          'two': pd.Series([15,47,89,63,52], index = ['a', 'c', 'd', 'e', 'f'])}
df = pd.DataFrame(mydata)
print(df)

imputer  = SimpleImputer(missing_values= np.nan, strategy = 'mean')
imputer.fit(df)

imputed_df = imputer.transform(df.values)
imputed_df = pd.DataFrame(imputer.transform(df.values))

imputed_df.columns = df.columns
imputed_df.index = df.index
print('---------------------------------------------\n')
print(imputed_df)
print('---------------------------------------------\n')
print(type(imputed_df))
```
